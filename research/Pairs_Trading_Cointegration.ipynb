```python
import pandas as pd # used for data analysis, cleaning and manipulation
import numpy as np # used primarily for calculations, because they are optimized for speed
import yfinance as yf # API to pull in financial data
import matplotlib.pyplot as plt # used for visualization purposes
import statsmodels.api as sm # used for varying statistical analyses
from statsmodels.tsa.stattools import coint # used for applying cointegration test for pair selection
from statsmodels.regression.linear_model import OLS # so we can do OLS
from statsmodels.tools.tools import add_constant  # so we can do OLS
```

```python
# upload the CSV that has S&P500 prices from 2017-01-01 to 2022-06-01
# and set the index column equal to the date
prices = pd.read_csv('SP500_Prices_17_22', index_col='Date')
prices.head()
#prices.shape
```

```python
def select_pairs(prices, start_date='2017-01-01', trade_start='2018-01-01'):
    """
    should add in start_date and end_date
    """
    train_prices = prices.loc[start_date:trade_start]  # 1 year formation period
    checked = []    # list to store the pairs that have been tested already
    cols = ['CADF p-value']      # values that we are interested in knowing for each pair
    pairs_info = pd.DataFrame(columns=cols)    # DataFrame to store info about the pairs
    
    for s1 in train_prices.columns:
        for s2 in train_prices.columns:
            #if len(checked) >= 5: # should make run time faster for just checking that code works
                #return pairs_info   # or break

            if s1 != s2 and (f'{s1} - {s2}' not in checked) and (f'{s2} - {s1}' not in checked):
                checked.append(f'{s1} - {s2}')
                cadf_pval = coint(train_prices[s1], train_prices[s2])[1]   # index 1 is the p-value
                if cadf_pval < 0.01:
                    pairs_info.loc[f'{s1} - {s2}'] = [cadf_pval]   # adds new row to dataframe with specified values
    return pairs_info
```

```python
def get_top_pairs(pairs_info, factor, direction=True, n_pairs=5):
    """
    pairs_info - DataFrame with information about the pairs
    factor - string with column name you want to sort by
    direciton - True for ascending and False for Descending
    n_pairs - number of pairs to return
    """
    top_5 = list(pairs_info.sort_values(by=factor, ascending=direction).index[:n_pairs]) 
    # returns a list of the pair names sorted based on factor and direction both specified
    return top_5
```

```python
def parse_pairs(pair):
    """
    split pair string "s1 - s2" and return the two tickers "s1" and "s2" of the pair
    """
    temp = pair.find(' - ')
    s1 = pair[:temp]
    s2 = pair[temp+3:]
    
    return s1, s2
```

```python
def plot_pairs(pairs, factor_info, start_date='2017-01-01', trade_start='2018-01-01', form_length=1, trade_length=6):
    """
    pairs - string of pairs to plot (in some order)
    factor_info - prices for 1 year formation period
    start_date - start date for formation period
    form_length - length of formation period in years, default to 1 year
    trade_length - length of trading period in months, default to 6 months
    
    change this to input one dataframe of info and start date for formation, then
    can create the 1 year formation data and 6 month trading data
    """
    
    for pair in pairs:
        s1, s2 = parse_pairs(pair)   # split pair to have strings for each stock individually
        
        
        form_end, trade_end = end_dates(start_date, form_length, trade_length)
        
        factor_train = factor_info.loc[start_date:form_end]
        
        
        factor_trade = factor_info.loc[trade_start:trade_end]
        
        res = OLS(factor_train[s1], add_constant(factor_train[s2])).fit() # this gets the hedge ratio
        hedge_ratio = res.params[s2]
        
        spread_form = factor_train[s1] - hedge_ratio*factor_train[s2]
        spread_trade = factor_trade[s1] - hedge_ratio*factor_trade[s2]
        spread_mean = spread_form.mean() # historical mean
        spread_std = spread_form.std()  # historical standard deviation
    
    
        fig, (ax1, ax2) = plt.subplots(1,2, figsize=(18,4))
        fig.suptitle(f'{pair}')
        ax1.set_title('Formation Period')
        ax1.plot(spread_form, label='spread')
        ax1.axhline(y=spread_mean, color='b', linestyle='dotted', label='mean')
        ax1.axhline(y=spread_mean+2*spread_std, color='r', linestyle='dotted', label='+2 SD')    
        ax1.axhline(y=spread_mean-2*spread_std, color='r', linestyle='dotted', label='-2 SD')
        ax1.legend()
        
        ax2.set_title('Trading Period')   
        ax2.plot(spread_trade, label='spread')
        ax2.axhline(y=spread_mean, color='b', linestyle='dotted', label='mean')
        ax2.axhline(y=spread_mean+2*spread_std, color='r', linestyle='dotted', label='+2 SD')    
        ax2.axhline(y=spread_mean-2*spread_std, color='r', linestyle='dotted', label='-2 SD')
        ax2.legend()
```

```python
def end_dates(start_date='2017-01-01', form_length=1, trade_length=6):
    """
    Given the start date, return the end date that is 1 year after
    form_length - length of formation period in years, default to 1 year
    trade_length - length of trading period in months, default to 6 months
    """
    form_end=list(start_date)
    start_year = ''.join(list(start_date)[0:4])
    form_end_year = str(int(start_year)+form__length)
    form_end[0:4]=list(form_end_year)
    form_end = ''.join(form_end)

    trade_end = list(form_end)
    trade_year = ''.join(list(trade_end)[0:4])
    trade_month = ''.join(list(trade_end)[5:7])
    #trade_end = form_end + 6 months
    if int(trade_month) > trade_length:
        # will have to shift year
        
        trade_year=str(int(trade_year)+1)
        trade_month=str(int(trade_month)-trade_length)
        trade_end[0:4] = list(trade_year)
        trade_end[5:7] = ['0'] + list(trade_month)
        trade_end = ''.join(trade_end)
    else:
        trade_month=str(int(trade_month)+trade_length)
        if int(trade_month) < 10: # if month is single digit, replace it with a zero infront
            trade_end[5:7] = ['0'] + list(trade_month)
        else:
            trade_end[5:7] = list(trade_month)
        trade_end = ''.join(trade_end)
    return form_end, trade_end
```

```python
#form_start = '2017-01-01'
#returns_num_total = 0
#returns_den_total = 0

#number_trade_periods = 2

# input: *start date* and number of trade periods
# 1. get formation period end and trade end dates
# 2. analyze all stocks over that formation period to select pairs
# 3. get the top 10 pairs to trade
# 4. calculate the profits and return variables for that trading period for those pairs
# 5. keep track of returns numerator and returns denominator
# 6. get next formation period start date
# 7. Loop


def run_strat(form_start = '2017-01-01', number_trade_periods=2, form_length=1, trade_length=6):
    returns_num_total = 0
    returns_den_total = 0
    for i in range(0, number_trade_periods):
        # get the dates for formation end (=trade start) and trade end
        form_end, trade_end = end_dates(form_start, form_length, trade_length)
        
        start_trade = form_end # the trade period start date is the formation period end date
        
        # anaylze all stocks over current formation period to select pairs
        
        # define current formation and trading periods
        form_current = prices.loc[form_start:form_end]
        trade_current = prices.loc[start_trade:trade_end]
        
        # select pairs based on current formation period - this is a DataFrame with statistical info about pairs (ex: CADF p-value)
        current_pairs = select_pairs(form_current, start_date=form_start, trade_start=start_trade)
        
        # extract top 10 pairs
        top_current_pairs = get_top_pairs(pairs_info=current_pairs, factor="CADF p-value", direction=True, n_pairs=10)
        
        # plot the pairs
        plot_pairs(top_current_pairs, prices, form_start, form_end)
        
        
        # calculate the returns (numerator and denominator) for all pairs in that trading period
        pair_profits, ret_num_tot, ret_den_tot, returns_vals, return_dates_current = returns_spread(top_current_pairs, prices, start_date=form_start, trade_start=start_trade)

        # keep track of returns (numerator and denominator)
        returns_num_total += ret_num_tot
        returns_den_total += ret_den_tot

        # get the dates for the next formation period start and end and trading period end

        form_start = get_next_start(trade_end = trade_end)


        if i == 0:
            pair_profits1 = pair_profits
            return_vals1 = returns_vals
        elif i == 1:
            pair_profits2 = pair_profits
            return_vals2 = returns_vals
    return pair_profits1, returns_num_total, returns_den_total, pair_profits2

    
    
```
